# Clean up Encoding Errors, Reduce to minimal expected cleanlieness 

```python
data = pd.read_csv("./Output/Gardenia_Unedited.csv",index_col=0)
```

```python
data.head()
```

```python
data.index = data.index.str.replace("\n","").str.capitalize()
```

```python
data.Hardiness = data.Hardiness.str.replace("What's My Zone\?","").str.strip()
data["Heat Zones"] = data["Heat Zones"].str.replace("What's My Zone\?","").str.strip()
data["Season of Interest"] = data["Season of Interest"].apply(lambda x: ")-".join(x.split(")")))
```

```python
for col in data:
    data[col] = data[col].str.replace(" – "," to ") 
```

```python
data.head()
```

```python
data.to_csv("./Output/Gardenia_Formatted.csv")
```

# Summaries 

```python
data = pd.read_csv("./Output/Gardenia_Formatted.csv",index_col=0)
```

Relevant Columns (in ERA) ["Sun Exposure","Soil Moisture","Plant Type","Pollinators","Flowering Months","Height (feet)","Showy"]

```python
full_df = None

for col in data:
    df = data[col].value_counts().reset_index()
    df.columns = [df.columns[1],"Count"]
    if full_df is not None:
        full_df = pd.concat([full_df,df],axis=1)
    else:
        full_df = df

full_df.to_csv("./Output/UniqueValues.csv")
```
